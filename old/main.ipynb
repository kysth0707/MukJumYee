텍스트 검출

```python
import cv2 as cv
import os
import matplotlib.pyplot as plt
```

```python
imagePos = r'E:\GithubProjects\MukJumYee\test\test.jpg'
```

```python
originalImage = cv.imread(imagePos)
# originalImage = cv.resize(originalImage, (1000, 1000))
```

```python
plt.imshow(originalImage)
```

```python
plt.imshow(cv.cvtColor(cv.cvtColor(originalImage, cv.COLOR_RGB2GRAY), cv.COLOR_BGR2RGB))
```

```python
imgCopy = originalImage.copy()
imgCopy = cv.cvtColor(imgCopy, cv.COLOR_RGB2GRAY)
_, binary130 = cv.threshold(imgCopy, 120, 255, cv.THRESH_BINARY)
plt.imshow(cv.cvtColor(binary130, cv.COLOR_BGR2RGB))
```

단순 binary threshold 는 특정 값을 기준으로 구분하기 때문에 환경 변화에 약하다  
그래서 otsu 알고리즘을 사용했다

```python
imgCopy = originalImage.copy()
imgCopy = cv.cvtColor(imgCopy, cv.COLOR_RGB2GRAY)
_, otsu = cv.threshold(imgCopy, -1, 255,  cv.THRESH_BINARY | cv.THRESH_OTSU)
plt.imshow(cv.cvtColor(otsu, cv.COLOR_BGR2RGB))
```
